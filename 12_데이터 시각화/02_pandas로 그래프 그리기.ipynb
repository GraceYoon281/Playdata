### pandas의 산점도

```python
import matplotlib.pyplot as plt
import pandas as pd

temperature = [25.2, 27.4, 22.9, 26.2, 29.5,
               33.1, 30.4, 36.1, 34.4, 29.1]
Ice_cream_sales = [236500, 357500, 203500, 365200, 
                   446600, 574200, 453200, 675400, 598400, 463100]
dict_data = {'기온':temperature, '아이스크림 판매량': Ice_cream_sales}
df_ice_cream = pd.DataFrame(dict_data, columns = ['기온', 
                                                  '아이스크림 판매량'])

df_ice_cream
```

```python
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df_ice_cream.plot.scatter(x='기온', y='아이스크림 판매량', grid=True, 
                          title='최고 기온과 아이스크림 판매량')
plt.show()
```

### pandas의 히스토그램

```python
import matplotlib.pyplot as plt
import pandas as pd

math = [76,82,84,83,90,86,85,92,72,71,100,
        87,81,76,94,78,81,60,79,69,74,87,82,68,79]

df_math = pd.DataFrame(math, columns = ['Student'])
math_hist = df_math.plot.hist(bins=8, grid= True)
math_hist.set_xlabel("시험 점수")
math_hist.set_ylabel("도수(frequency)")
math_hist.set_title("수학 시험의 히스토그램")
```

### pandas의 파이 그래프

```python
import matplotlib.pyplot as plt

import pandas as pd

fruit = ['사과', '바나나','딸기','오렌지','포도']
result =[7, 6, 3, 2, 2]

df_fruit = pd.Series(result, index = fruit, name ='선택한 학생수')
df_fruit
```

```python
df_fruit.plot.pie()
plt.show()
```

테클 안걸릴때까지 하는 거 누가 보던 간에 잘했다 할 수 있도록


파이로 이동: 정리 해놓는 것 노션에 

```python
explode_value = (0.1, 0, 0, 0, 0)
fruit_pie = df_fruit.plot.pie(figsize = (5, 5), autopct = '%.1f%%', startangle=90,
                             counterclock = False, explode = explode_value, shadow = True, table = True)
fruit_pie.set_ylabel("")
fruit_pie.set_title("과일 선호도 조사 결과")

plt.savefig('c:/mypycode/figures/saveFigTest3.png', dpi = 200)
plt.show()
```
