```python
# 놀이 동산
'''
키가 100cm 이상이고, 나이가 5세 이상이면 놀이 기구 탑승 가능

'''

height = int(input("아이의 키를 입력하세요, 숫자로"))
age = int(input("아이의 나이를 입력하세요, 숫자로"))

if height >= 100 and age >= 5:
    print("탑승 가능 합니다")

else:
    print("탑승 불가능 합니다")
```
