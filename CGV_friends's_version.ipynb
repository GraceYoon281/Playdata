```python
# 문지현's version

day1 = ["월요일","월","화요일","화","수요일","수","목요일","목","금요일","금"]
day2 = ["일요일","일","토요일","토"]
day = day1 + day2 


while True:                                                           #예외처리 하는 부분
    try:
        age = int(input("나이를 입력해주세요 : "))
        break
    except ValueError:
        print("나이를 다시 입력해주세요")

        
while True:
    try:
        movie = input("2D 와 3D관 관람하실 관을 선택해주세요 : ")
        if movie == "2D":
            print(movie + "관을 선택하셨습니다.")
            break
        elif movie == "3D":
            print(movie + "관을 선택하셨습니다.")
            break    
        else:
            print("오류입니다.")
    except ValueError:
        print("상영관을 다시 선택해주세요")

        
while True:
    try:    
        dayinput = input("요일을 입력해주세요 : ")
        if dayinput in day :
            break  
        else :
            print("오류입니다")  
    except ValueError:
            print("요일을 다시 입력해주세요")
            
            
money = 12000;
if age <= 19:
    money = money - 2000
else : money = money
    
if "3D" in movie:
    money = money + 3000
else : money = money
    
if dayinput in day2:
    money = money + 2000
else : money = money

print("고객님의 요금은 "+str(money)+" 입니다.")

```

```python
# 문지현`s version 2

day1 = ["월","화","수","목","금"]
day2 = ["일","토"]
day = day1 + day2 

while True:                                                           #예외처리 하는 부분
    try:
        age = int(input("나이를 입력해주세요 : "))
        break
    except ValueError:
        print("나이를 다시 입력해주세요")
        
while True:
    try:
        movie = input("2D 와 3D관 관람하실 관을 선택해주세요 : ")
        if movie == "2D":
            print(movie + "관을 선택하셨습니다.")
            break
        elif movie == "3D":
            print(movie + "관을 선택하셨습니다.")
            break    
        else:
            print("오류입니다.")
    except ValueError:
        print("상영관을 다시 선택해주세요")

while True:
    try:    
        dayinput = input("요일을 입력해주세요 : ")     
        if "요일" in dayinput:
            dayinput = dayinput.replace("요일","")
            break
        elif dayinput in day:
            dayinput = str(dayinput)
            break
        else :
            print("오류입니다")  
    except ValueError:
            print("요일을 다시 입력해주세요")

            
money = 12000;
if age <= 19:
    money = money - 2000
else : money = money
    
if "3D" in movie:
    money = money + 3000
else : money = money
    
if dayinput in day2:
    money = money + 2000
else : money = money

print("고객님의 요금은 "+str(money)+" 입니다.")
```
